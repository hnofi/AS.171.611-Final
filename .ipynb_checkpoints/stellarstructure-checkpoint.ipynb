```python
# Constants in CGS units
sigma_sb = 5.67e-5
Na = 6.02e23
Rg = 8.314
c = 2.99e10
G = 6.67e-8
k = 1.38e-16
a = 7.56e-15

Ms = 1.989e33
Rs = 6.96e10
Ls = 3.826e33
```

```python
def opacity(lines, rt, Tc):
    if rt <= 0 or Tc <= 0:
        return 1e-4  # or some default small opacity value
    logR = np.array([float(x) for x in lines[0].split()[2:]])
    data = np.loadtxt(lines[2:])  
    logT = data[:, 0]
    kappa = data[:, 1:]
    interpolator = RegularGridInterpolator((logT, logR), kappa, bounds_error=False, fill_value=None)
    k = interpolator((np.log10(Tc), np.log10(rt)))
    return 10**k
```

```python
def energy_generation_pp(rho, T, X):
    #KWW 18.51 equ 18.63
    psi = 1 #?
    T9 = T/1e9 
    T7 = T/1e7
    
    #f_11 equation from google
    t = 5.92e-3*1*((rho/(T7**3))**(1/2))
    f_11 =  np.exp(t)
    
    #g_11
    g_11 = 1 + (3.82* T9) + (1.51* T9**2) + (0.144* T9**3) - (0.0114* T9**4)
    
    #epsilon
    ep_pp = 2.57e4 * psi * f_11 * g_11 * rho * X**2 * T9**(-2/3) * np.exp(-3.381/(T9**(1/3)))
    return ep_pp
```

```python
def energy_generation_CNO(rho, T, X, Z):
    #KWW 18.51 equ 18.65
    T9 = T/1e9
    
    X_cno = Z #assuming most metals are carbon, oxygen, and nitrogen
    
    g_141 = 1 - (2.00*T9) + (3.41*T9**2) - (2.43*T9**3)
    
    ep_CNO = 8.24e25*g_141*X_cno*X*rho*T9**(-2/3)*np.exp((-15.231*T9**(-1/3))-(T9/0.8)**2)
    return ep_CNO
```

```python
def density(X, Y, T, P):
    N_A = 6.023e23
    k = 1.38e-16 
    a = 7.56e-15

    Pr = ((1/3)*a*T**4)
    Pg = P - Pr
    if Pg <= 0:
        return np.nan  # to trigger a fail before log10 is taken

    Z = 1 - X - Y
    mu = 4 / (3 + 5*X)

    rho = (Pg * mu) / (N_A * k * T)
    return rho
```

```python
def load1(Tc, Pc, X, Z, l,op_table, m, m0=1e-5):
    """
    Initial conditions at a small mass m0 near the center of the star.
    Assumes:
    - l = 0 (no luminosity at center)
    - P = Pc, T = Tc (central values)
    - r ≈ small nonzero seed (say 10 km) to avoid singularity
    """
    #l = 0.0
    #P = Pc
    #T = Tc
    #r = 1e7  # small non-zero radius (10 km)
    Mcp = 1e-10 * m
    a = 7.56*10**(-15)
    rho = density(X, Y, Tc,Pc)
    #print(rho)
    
    # l: KWW equ 11.4
    e_pp = energy_generation_pp(rho, Tc, X)
    e_cno = energy_generation_CNO(rho, Tc, X, Z)
    l = (e_pp+e_cno) * Mcp
    print(l)
    
    # P: KWW equ 11.6
    P = Pc - (3*G/(8*np.pi))*((4*np.pi*rho/3)**(4/3))*(Mcp**(2/3))
    #print(P)
    
    # r: KWW equ 11.3
    r = (3*Mcp/(4*np.pi*rho))**(1/3)
    print(r)

    # T: KWW equ 11.9 radiative
    rt = rho / Tc**3
    kappa = opacity(op_table, rt, Tc)
    T = (Tc**4 - 1/(2*a*c) * (3/(4*np.pi))**(2/3) * kappa * (e_pp + e_cno) * rho**(4/3) * Mcp**(2/3))**(1/4) 
    return np.array([l, P, r, T])
```

```python
def load2(m, L_star, R_star, Pc, Tc, X, Y, Z, op_table):
    """
    Initial conditions at the surface of the star (mass = M, radius = R_star).
    Uses:
    - L_star = total luminosity
    - R_star = total radius
    - T from Stefan-Boltzmann
    - P = small (photospheric) pressure
    """
    l = L_star
    r = R_star
    
    T_eff = (L_star / (4*np.pi*R_star**2*sigma_sb))**(1/4)
    
    rho = density(X, Y, T_eff, Pc)
    rt = rho / T_eff**3
    kappa = opacity(op_table, rt, T_eff)

    #print(kappa)
    P = (2*G*m/(3*r**2*kappa))
    #print(P)
    return np.array([l, P, r, T_eff])
```

```python
def derivs(m, y, X, Y, Z, op_table):
    l, P, r, T = y  # unpack dependent variables
    rho = density(X, Y, T, P)
    
    # Check for non-physical values to prevent crashing
    if rho <= 0 or T <= 0:
        return [0, 0, 0, 0]

    e_pp = energy_generation_pp(rho, T, X)
    e_cno = energy_generation_CNO(rho, T, X, Z)
    kappa = opacity(op_table, np.log10(rho), np.log10(T))
    
    a = 7.56e-15
    c = 2.998e10
    G = 6.67259e-8
    
    del_rad = 3 * P * l * kappa / (16 * np.pi * a * c * G * m * T**4)
    del_ad = 0.4  # or 5/3 - 1 depending on γ
    
    del_ = min(del_rad, del_ad)

    dl_dm = e_pp + e_cno
    dP_dm = -G * m / (4 * np.pi * r**4)
    dr_dm = 1 / (4 * np.pi * r**2 * rho)
    dT_dm = -G * m * T / (4 * np.pi * r**4 * P) * del_
    
    return [dl_dm, dP_dm, dr_dm, dT_dm]
```

```python
def load1(Tc, Pc, X, Z, l,op_table, m, m0=1e-5):
    """
    Initial conditions at a small mass m0 near the center of the star.
    Assumes:
    - l = 0 (no luminosity at center)
    - P = Pc, T = Tc (central values)
    - r ≈ small nonzero seed (say 10 km) to avoid singularity
    """
    l = 0.0
    P = Pc
    T = Tc
    r = 1e7  # small non-zero radius (10 km)
    return np.array([l, P, r, T])
```

```python
def load2(m, L_star, R_star, Pc, Tc, X, Y, Z, op_table):
    """
    Initial conditions at the surface of the star (mass = M, radius = R_star).
    Uses:
    - L_star = total luminosity
    - R_star = total radius
    - T from Stefan-Boltzmann
    - P = small (photospheric) pressure
    """
    sigma = 5.67051e-5
    l = L_star
    r = R_star
    T = (L_star / (4 * np.pi * R_star**2 * sigma))**(1/4)
    P = 1e5  # surface pressure ~ photosphere
    return np.array([l, P, r, T])
```

```python
def shootf(m, L_star, R_star, Pc, Tc, X, Y, Z, op_table, detailed=False, M_meeting_point = 0.5,M_start=1e-5):
    M = m
    args = (X, Y, Z, op_table)
    steps = 10000
    Mr_array_from_centre = np.linspace(M_start, M_meeting_point, steps) * M
    sol_from_centre = scipy.integrate.solve_ivp(derivs, (Mr_array_from_centre[0], Mr_array_from_centre[-1]), load1(Tc, Pc, X, Z, L_star, op_table, m), args=args, method='RK45', t_eval=Mr_array_from_centre)
    Mr_array_from_surface = np.linspace(1, M_meeting_point, steps) * M
    sol_from_surface = scipy.integrate.solve_ivp(derivs, (Mr_array_from_surface[0], Mr_array_from_surface[-1]), load2(m, L_star, R_star, Pc, Tc, X, Y, Z, op_table), args=args, method='RK45', t_eval=Mr_array_from_surface)
    return sol_from_centre, sol_from_surface
```

```python
Ms = 1.988e33  # g
mass_mesa = profile.mass # solar masses
pp = profile.pp
cno = profile.cno
eps = pp + cno  # erg/g/
mass_mesa = mass_mesa[::-1]
eps = eps[::-1]
# Compute dm (shell mass) in grams
dm = np.empty_like(mass_mesa)
dm[0] = mass_mesa[0]
dm[1:] = mass_mesa[1:] - mass_mesa[:-1]
dm *= Ms  # now in grams

# Integrate luminosity from center out
luminosity = np.cumsum(eps * dm)  # erg/s

# Total stellar luminosity:
L_total = luminosity[-1]
print(f"Total luminosity = {L_total:.3e} erg/s")
print(f"In solar units = {L_total / 3.828e33:.3f} Lsun")
```
