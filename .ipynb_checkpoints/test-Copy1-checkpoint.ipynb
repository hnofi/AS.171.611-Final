```python
import numpy as np
from scipy.interpolate import RegularGridInterpolator
import matplotlib.pyplot as plt
from astropy import units as u
import scipy

# Constants in CGS units
sigma_sb = 5.67051e-5
Na = 6.0221415e23
Rg = 8.314
c = 2.99792458e10
G = 6.67259e-8
k = 1.380658e-16
a = 7.5646e-15

Ms = 1.989e33
Rs = 6.96e10
Ls = 3.826e33

with open('LosAlamosY=0.28.txt', 'r') as f:
    lines = f.readlines()

def opacity(lines, rt, Tc):
    if rt <= 0 or Tc <= 0:
        return 1  # or some default small opacity value
    logR = np.array([float(x) for x in lines[0].split()[2:]])
    data = np.loadtxt(lines[2:])  
    logT = data[:, 0]
    kappa = data[:, 1:]
    interpolator = RegularGridInterpolator((logT, logR), kappa, bounds_error=False, fill_value=None)
    k = interpolator((np.log10(Tc), np.log10(rt)))
    return 10**k

def energy_generation_pp(rho, T, X):
    #KWW 18.51 equ 18.63
    psi = 1 #?
    T9 = T/1e9 
    T7 = T/1e7
    
    #f_11 equation from google
    t = 5.92e-3*1*((rho/(T7**3))**(1/2))
    f_11 =  np.exp(t)
    
    #g_11
    g_11 = 1 + (3.82* T9) + (1.51* T9**2) + (0.144* T9**3) - (0.0114* T9**4)
    
    #epsilon
    ep_pp = 2.57e4 * psi * f_11 * g_11 * rho * X**2 * T9**(-2/3) * np.exp(-3.381/(T9**(1/3)))
    return ep_pp

def energy_generation_CNO(rho, T, X, Z):
    #KWW 18.51 equ 18.65
    T9 = T/1e9
    
    X_cno = Z #assuming most metals are carbon, oxygen, and nitrogen
    
    g_141 = 1 - (2.00*T9) + (3.41*T9**2) - (2.43*T9**3)
    
    ep_CNO = 8.24e25*g_141*X_cno*X*rho*T9**(-2/3)*np.exp((-15.231*T9**(-1/3))-(T9/0.8)**2)
    return ep_CNO

def get_epsilon_energy_generation(rho, T, X, Z):
    epsilon_pp = energy_generation_pp(rho, T, X)
    epsilon_cno = energy_generation_CNO(rho, T, X, Z)
    epsilon = epsilon_pp + epsilon_cno
    return epsilon

def density(X, Y, T, P):
    N_A = 6.023e23
    k = 1.38e-16 
    a = 7.56e-15

    Pr = ((1/3)*a*T**4)
    Pg = P - Pr
    if Pg <= 0:
        return np.nan  # to trigger a fail before log10 is taken

    Z = 1 - X - Y
    mu = 4 / (3 + 5*X)

    rho = (Pg * mu) / (N_A * k * T)
    return rho
def load1(Tc, Pc, X, Y, Z, lines, m):
    l = 0.0
    P = Pc
    T = Tc
    r = 1e7  # small non-zero radius (10 km)
    return np.array([l, P, r, T])


def load2(m, L_star, R_star, Pc, Tc, X, Y, Z, op_table): 
    """
    l, P , r, and T at the surface
    """
    sigma = 5.67051e-5
    l = L_star
    r = R_star
    T = (L_star / (4 * np.pi * R_star**2 * sigma))**(1/4)
    P = 1e5  # surface pressure ~ photosphere
    return l, P, r, T

def derivs(m, y, X, Y, Z, op_table):
    l, P, r, T = y
    rho = density(X, Y, T, P)
    
    # Check for non-physical values to prevent crashing
    if rho <= 0 or T <= 0:
        return [0, 0, 0, 0]

    e_pp = energy_generation_pp(rho, T, X)
    e_cno = energy_generation_CNO(rho, T, X, Z)
    kappa = opacity(op_table, np.log10(rho), np.log10(T))
    
    a = 7.56e-15
    c = 2.998e10
    G = 6.67259e-8
    
    del_rad = 3 * P * l * kappa / (16 * np.pi * a * c * G * m * T**4)
    del_ad = 0.4  # or 5/3 - 1 depending on Î³
    
    del_ = min(del_rad, del_ad)

    dl_dm = e_pp + e_cno
    dP_dm = -G * m / (4 * np.pi * r**4)
    dr_dm = 1 / (4 * np.pi * r**2 * rho)
    dT_dm = -G * m * T / (4 * np.pi * r**4 * P) * del_
    
    return [dl_dm, dP_dm, dr_dm, dT_dm]


def shootf(m, L_star, R_star, Pc, Tc, X, Y, Z, op_table, detailed=False, M_meeting_point = 0.5
,M_start=1e-5):
    M = m
    args = [{'M': M, 'mu': 0.6}]
    steps = 100 if not detailed else 10000
    m0 = M_start * M
    Mr_array_from_centre = np.linspace(m0, M_meeting_point * M, steps)
    sol_from_centre = solve_ivp(
    derivs,
    t_span=(Mr_array_from_centre[0], Mr_array_from_centre[-1]),
    y0=load1(Tc, Pc, X, Y, Z, op_table, m0),
    args=(X, Y, Z, op_table),
    method='RK45',
    t_eval=Mr_array_from_centre
)
    Mr_array_from_surface = np.linspace(1, M_meeting_point, steps) * M
    sol_from_surface = solve_ivp(
    derivs,
    t_span=(Mr_array_from_surface[0], Mr_array_from_surface[-1]),
    y0=load2(M, L_star, R_star, Pc, Tc, X, Y, Z, op_table),
    args=(X, Y, Z, op_table),
    method='RK45',
    t_eval=Mr_array_from_surface
)
    return sol_from_centre, sol_from_surface


M = 2.2 * Ms
X = 0.7
Y = 0.28
Z = 0.02

mu = 4 / (3 + 5*X)

R = Rs * (M / Ms)**0.75
L = Ls * (M / Ms)**3.5

# constant density model
Pc = 3 / (8 * np.pi) * G * M**2 / R**4
Tc = 1/2 * G * M / R * mu / (Na * k)

opacity_interp = opacity(lines, density(X,Y,Tc, Pc), Tc)

args = (X, Y, Z, lines)

args

initial_guess = np.array([Pc, R, L, Tc])
initial_guess

sol1, sol2 = shootf(M, L, R, Pc, Tc, X, Y, Z, lines, detailed=True, M_start=1e-5)
```
